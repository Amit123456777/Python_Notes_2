```python
import pandas as pd
resultsheet={'jack':pd.Series([91,92,93],index=['maths','science','english']),
            'john':pd.Series([82,83,84],index=['maths','science','english']),
            'jill':pd.Series([92,74,58],index=['maths','science','english']),}
RDF=pd.DataFrame(resultsheet)
print(RDF)
```

```python
# ACCESS ELEMENTS FROM DATA FRAME
```

```python
# add the person 
RDF['harsh']=[45,76,98]
print(RDF)
```

```python
RDF.loc['accountance']=[32,65,97,98]
print(RDF)
```

```python
# for deleting the row
# to delete a row axis value is 0 and for coloum deleting value is 1
RDF=RDF.drop('science',axis=0)
print(RDF)
```

```python
# for deleting the coloum
RDF=RDF.drop('jack',axis=1)
print(RDF)
```

```python
import pandas as pd

dFrame1 = pd.DataFrame([[1, 2, 3], [4, 5], [6]],
columns=["c1", "c2", "c3"],
index=["r1", "r2", "r3"])
print(dFrame1)

dFrame2 = pd.DataFrame([[10, 20, 30], [40, 50], [60]],
columns=["c1", "c2", "c3"],
index=["r1", "r2", "r3"])
print(dFrame2)

dFrame = dFrame1._append(dFrame2)
print(dFrame)
```
