```python
# Sorting

def is_sorted(arr):
    sorted = True
    for i in range(len(arr)-1):
        if arr[i]>arr[i+1]:
            sorted = False
            
    return sorted
arr = [9,2,50,74,1]
is_sorted(arr)
```

```python
# Monkey sort

import random

def is_sorted(arr):
    for i in range(1, len(arr)):
        if arr[i] < arr[i - 1]:
            return False
    return True

def monkey_sort(arr):
    while not is_sorted(arr):
        random.shuffle(arr)
        print(arr)
    print(arr)

a = [23, 14, 11, 67, 45, 61]
monkey_sort(a)


# is_sorted is a function used to 
# check whether a given list is sorted in ascending order.
# It returns True if the list is sorted and False otherwise.
```
