```python
# write a program to draw a scatter plot comparing two subject marks of
# mathematics  and science.
```

```python
import matplotlib.pyplot as plt

range_marks = [10, 20, 30, 40, 50]
math_marks = [88, 97, 96, 92, 85]
science_marks = [99, 97, 56, 78, 94]

plt.scatter(range_marks, math_marks, label="math_marks")
plt.scatter(range_marks, science_marks, label="science_marks")
plt.legend()
plt.xlabel("Range")
plt.ylabel("Marks")
plt.title("Math and Science Marks")
plt.show()
```

# Pie Chart

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = [33, 52, 12, 17,42,48]
plt.pie(data)
plt.show()
```

# adding labels

```python
label = ["python", "c","c++", "c" , "php", "javascript"]
plt.pie(data,labels= label)
plt.show()
```

```python
plt.figure(figsize = (9,7))
plt.pie(data)
plt.title("the result of the programming language survey: ", fontsize =15)
plt.legend(label)
plt.show()
```

```python
# explode
explode = [0.0, 0.08, 0.10, 0.0, 0.00,0.00]

plt.pie(data, explode = explode, labels = label)
plt.show()
```

```python
# Donut chart
import matplotlib.pyplot as plt
data = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
labels = ["python", "c", "c++", "php", "java","ruby"]
plt.pie(data, labels = labels, startangle= 90, wedgeprops = {"width" : 0.2})
plt.show()
```

```python
# contour plot
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate some Data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis')  # Corrected 'ac' to 'ax' and 'cmaps' to 'cmap'

# ADD labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Plot of $sin(\\sqrt{X^2 + Y^2})$')  # Corrected the title string

# Add color bar
fig.colorbar(surf)

# Show plot
plt.show()
```
